```python
'''
문제) 사용자에게 연도를 입력으로 받아 윤년 여부를 출력하는 코드를 작성하세요.

실행 예1)                    실행 예2)
년을 입력하세요: 2023        년을 입력하세요: 2020
2023년은 평년입니다.         2020년은 윤년입니다.

[윤년 판단 알고리즘]
1. 4의 배수인 해는 윤년입니다.
2. 100의 배수인 해는 평년입니다.
    ==> 1-2. 4의 배수이면서 100의 배수가 아닌 년도는 윤년입니다.
3. 400의 배수인 해는 윤년입니다.
    => 2022: 평년, 2020: 윤년, 2100: 평년, 2000: 윤년
 
 [문제풀이 유형]
 1. if ... elif ... else 구문을 사용하여 문제를 해결합니다.
 2. if ... else 구문을 사용하여 문제를 해결합니다.
'''`
```

```python
# 1. if ... elif ... else 구문을 사용하여 문제를 해결합니다.

year = int(input('년도를 입력하세요: '))

if year % 400 == 0: # 상위 조건은 가장 범위가 좁아야 함, overwrite 될 수 있는 조건부터
    result = '윤년'
elif year % 100 == 0:
    result = '평년'
elif year % 4 == 0:
    result = '윤년'
else:
    result = '평년'

print(str(year) + '년은 ' + str(result) + '입니다.')
```

```python
#  2. if ... else 구문을 사용하여 문제를 해결합니다.

year = int(input('년도를 입력하세요: '))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0): # 알고리즘을 두 가지 경우로 묶기 위해 2번을 윤년 조건으로 바꿔야 함
    result = '윤년'
else:
    result = '평년'

print(str(year) + '년은 ' + str(result) + '입니다.')

# if와 else만으로 구분되므로 모두 윤년 조건으로 만들기 위해 1, 2번 조건을 합치거나 2, 3번 조건을 합쳐야 함.
# 조건을 합칠 때는 더 넓은 범위의 것과 합쳐야 함으로 1, 2번을 합쳐야 함.
```

```python
# answer, for문

total = 0

for num in range(1,101):
    total = total + num

print('1부터 100까지의 합: ', total)
```

```python
# answer, while문

total = 0
num = 1 # for 반복문 변수를 밖으로 빼고 범위의 시작값으로 설정

while num < 101: # 조건 생성
    total = total + num # while 문 안에서 num 값을 바꿔줘야 함. 그래야 끝이 남.
    num = num + 1 # 반복할 때마다 num 값은 1씩 증가해야 함

print('1부터 100까지의 합: ', total)

# while 문은 반드시 while 문 안에서 조건의 미지수를 변경해줘야 함.
```

```python
'''
문제3) 다음 그림과 같이 제목이 있는 수평 구구단 결과를 출력하는 코드를 작성하세요.
        단, 하나의 중첩 for 반복문을 사용합니다.
'''
```

```python
# answer

num = 0
while num < 10:
    for dan in range(2, 10):
        if num == 0:
            print('[ %d 단 ]' % (dan), end = '\t')
        else:
            print('%d * %d = %2d' % (dan, num, dan * num), end='\t')
    
    print()
    num = num + 1
```

```python
# while 중첩

num = 0
while num < 10:
    dan = 2
    while dan < 10:
        if num == 0:
            print('[ %d 단 ]' % (dan), end = '\t')
        else:
            print('%d * %d = %2d' % (dan, num, dan * num), end='\t') # \t = tab
        dan = dan + 1 # dan이 속한 while 안에서 변경
        
    print()
    num = num + 1
```

```python
# answer

import random

kbb = ['가위', '바위', '보']
intro = '''
    [ 가위바위보 게임 ]
    
    가위바위보를 선택하세요!
    1. 가위
    2. 바위
    3. 보
    >>> '''

is_win = False

while not is_win:
    com = random.randint(0, 2)
    user = int(input(intro)) - 1
    
    if user == com:
        result = '비겼습니다.'
    elif (user == 1 and com == 0) or (user == 2 and com == 1) or (user == 0 and com == 2):
        result = '이겼습니다.'
        is_win = True
    else:
        result = '졌습니다.'

    print('\n    사용자:', kbb[user], ', 컴퓨터:', kbb[com])
    print('    결과: 사용자가', result)
    
print('\n\n    [ 게임을 종료합니다. ]')
```

```python
# list-for comprehension

lst1 = [1, 2, 3, 4, 5]
lst2 = [num ** 2 for num in lst1] # for 헤더 대괄호에 넣고 앞에 관련 연산 붙이기
print(lst2)
```

```python
# set-for comprehension

set1 = {1, 2, 3, 4, 5}
set2 = {num ** 2 for num in lst1}
print(set2)
```

```python
# dict-for comprehension

dct1 = {'one' : 1, 'two' : 2, 'three' : 3}
dct2 = {key:value ** 2 for key, value in dct1.items()}
print(dct2)
```

```python
'''
문제2) 사용자에게 년, 월을 입력으로 받아 입력 받은 해의 해당 월의 마지막 날의 일자를 출력하는 코드를 작성하세ㅛㅇ.

실행 예1)                                  실행 예2)

년을 입력하세요: 2022                      년을 입력하세요: 2020
월을 입력하세요: 9                         월을 입력하세요: 2

2022년 9월의 마지막 일은 30일 입니다.      2022년 2월의 마지막 일은 29일 입니다.

[월의 마지막 일자 판단 알고리즘]

1. 큰 달 / 1, 3, 5, 7, 8, 10, 12 => 31일
2. 작은 달 / 4, 6, 9, 11 => 30일
3. 2일 / 윤년 => 29일, 평년 => 28일

'''
```

```python
# my

year = int(input('년을 입력하세요: '))
mon = int(input('월을 입력하세요: '))


if mon == 2:
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        result = 29
    else:
        result = 28
elif mon == 4 or 6 or 9 or 11:
        result = 30
elif mon == 1 or 3 or 5 or 7 or 8 or 10 or 12:
        result = 31
    
print('2022년 %d월의 마지막 일은 %d일 입니다.' % (mon, result))
```
