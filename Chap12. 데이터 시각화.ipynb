```python
# 수평에서는 width 지정 불가
x = np.arange(len(member_IDs))
colors = ['r', 'g', 'b', 'm']
plt.barh(x, before_ex, tick_label = member_IDs, color = colors)

plt.show()
```

```python
# 중첩 막대그래프
x = np.arange(len(member_IDs))
plt.bar(x, before_ex, color = 'c', align='edge', width=0.4, label='before')
plt.bar(x+0.4, after_ex, color = 'm', align='edge', width=0.4, label='after')
plt.xticks(x+0.4, member_IDs)
plt.legend()
plt.xlabel('회원 ID')
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 시작 전후의 근지구력(복근) 변화 비교')

plt.show()
```

```python
x = np.arange(len(member_IDs))
bar_width = 0.4
plt.bar(x, before_ex, color = 'c', align='edge', width=0.4)
plt.bar(x + bar_width, after_ex, color = 'm', align='edge', width=0.4)
plt.xticks(x + bar_width, member_IDs)
plt.legend(['before', 'after'])
plt.xlabel('회원 ID')
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 시작 전후의 근지구력(복근) 변화 비교')

plt.show()
```
