```python
from bs4 import BeautifulSoup
```

```python
html = '''
<html>
    <body>
        <div>
            <span>
                <a href='http://www.naver.com'>naver</a>
                <a href='https://www.google'>google</a>
                <a href='http://www.daum.net'>daum</a>
            </span>
        </div>
    </body>
</html>
'''
```

```python
soup = BeautifulSoup(html, 'lxml')
print(soup)
```

```python
tag = soup.find('a') # find의 결과는 tag 객체

print(type(tag), tag)
```

```python
tag.get_text() # 시작과 종료 태그 사이에 있는 텍스트
```

```python
tags = soup.find_all('a') # find_all의 결과는 result set(=list)

print(type(tags), tags) # 3개의 a tag에 대한 객체가 들어감
```

```python
url = 'https://finance.naver.com/sise/sise_market_sum.nhn'
html = requests.get(url).text
soup = BeautifulSoup(html, 'lxml')

top = soup.select('table.type_2>tbody>tr')
top_50 = []

for i in range(len(top)):
    if len(top[i].select('td.no')) == 1:
        top_50.append(top[i])
        
len(top_50)
```

```python
top_50[0]
```

```python
ranking = top_50[0].select('td.no')
ranking[0].get_text()
```

```python
tltle = top_50[0].select('td>a.tltle')
tltle[0].get_text()
```

```python
curr_price = top_50[0].select('td.number')
curr_price[0].get_text()
```

```python
top_stock = {'순위':[], '종목명':[], '현재가':[]}

for stock in top_50:
    print(stock.select('td.no')[0].get_text(), end='\t')
    print(stock.select('td>a.tltle')[0].get_text(), end='\t')
    print(stock.select('td.number')[0].get_text())
```

```python
top_stock = {'순위':[], '종목명':[], '현재가':[]}

for stock in top_50:
    top_stock['순위'].append(stock.select('td.no')[0].get_text())
    top_stock['종목명'].append(stock.select('td>a.tltle')[0].get_text())
    top_stock['현재가'].append(stock.select('td.number')[0].get_text())
```

```python
import pandas as pd
```

```python
top_stock_df = pd.DataFrame(top_stock, columns=['종목명', '현재가'], index=top_stock['순위'])
top_stock_df.index_name='순위'
top_stock_df
```

```python
url = 'https://finance.naver.com/sise/sise_market_sum.nhn'
html = requests.get(url).text
soup = BeautifulSoup(html, 'lxml')

top = soup.select('table.type_2>tbody>tr')
top_50 = []

for i in range(len(top)):
    if len(top[i].select('td.no')) == 1:
        top_50.append(top[i])
        
top_stock = {'순위': [], '종목명': [], '현재가': []}

for stock in top_50:
    top_stock['순위'].append(stock.select('td.no')[0].get_text())
    top_stock['종목명'].append(stock.select('td>a.tltle')[0].get_text())
    top_stock['현재가'].append(stock.select('td.number')[0].get_text())
    
top_stock_df = pd.DataFrame(top_stock, columns=['종목명', '현재가'], index=top_stock['순위'])
top_stock_df.index_name='순위'
top_stock_df
```
